```python
canciones = ["a", "b", "c", "d", "e"]
duraciones = [2.1, 3.4, 5.3, 1.2, 0.2]

diccionario = dict(zip(canciones, duraciones))
diccionario
```

```python
masLargas = []
diccionario_ordenado = dict(sorted(diccionario.items(), key=  lambda x: x[1], reverse = True))
for i in [0,1,2]: 
    masLargas.append(list(diccionario_ordenado.items())[i])


print(masLargas)
```
