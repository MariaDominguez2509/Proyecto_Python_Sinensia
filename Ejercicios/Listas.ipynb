### Primer ejercicio

```python
top_ventas = [
    "Thriller",
    "Back in Black",
    "The Dark Side of the Moon",
    "The Bodyguard",
    "Rumours"
]
```

```python
top_ventas.append("Abbey Road")
top_ventas.pop()
top_ventas 
```

```python
top_ventas.insert(1, "Abbey Road")
top_ventas 
```

```python
indice = top_ventas.index("The Bodyguard")
top_ventas.pop(indice)
top_ventas
```

```python
top_ventas.reverse()
top_ventas
```

```python
top_ventas[0]
```

```python
for a in top_ventas.copy(): 
    top_ventas.remove(a)

top_ventas
```
